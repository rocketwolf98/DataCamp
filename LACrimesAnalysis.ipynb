```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
# Start coding here
# Use as many cells as you need
# Determine the highest number of crimes

# Ensure 'TIME OCC' is treated as a string
crimes['TIME OCC'] = crimes['TIME OCC'].astype(str)

# Extract the hour from 'TIME OCC'
crimes['HOUR OCC'] = crimes['TIME OCC'].str[:2]

# Determine the peak crime hour
crime_counts = crimes['TIME OCC'].str[:2].value_counts().index[0]
peak_crime_hour = int(crime_counts)
print(f'The peak crime hour is: {peak_crime_hour}')
```

```python
# Treat times as integer values
crimes['TIME OCC'] = crimes['TIME OCC'].astype(int)

# Create a list of hours to retain (22:00 to 23:59 and 00:00 to 03:59)
hours_to_retain = list(range(2200, 2400)) + list(range(0, 400))

# Filter the DataFrame using boolean indexing and .isin() method
crimes_filter = crimes[crimes['TIME OCC'].isin(hours_to_retain)]

# Revert 'TIME OCC' to string
crimes_filter['TIME OCC'] = crimes_filter['TIME OCC'].astype(str)

# Group by location and count occurrences
loc_grouped = crimes_filter.groupby('AREA NAME').size().reset_index(name='count')

# Determine the location with the highest number of crimes
peak_night_crime_location = loc_grouped.loc[loc_grouped['count'].idxmax(), 'AREA NAME']

peak_night_crime_location
```
